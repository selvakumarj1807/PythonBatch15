# 1. Arithmatic Operators

```python
a = 7

b = 2

print('Addition (+) : ', a + b)

print('Subtraction (-) : ', a - b)

print('Multiplication (*) : ', a * b)

print('Division (/) : ', a / b) # It will return Quotient

print('-----------------------------')

print('Modulas (%) : ', a % b) # It will return Remainder

print('Floor Division (//) : ', a // b)

print('Exponentiation (**) : ', a ** b) # 7 ** 2 = 7 * 7 = 49
```

# 3. Comparision or Relational Operators

```python
a = 10

b = 5

print('Equal (==) : ', a == b)

print('Not Equal (!=) : ', a != b)

print('Greater than (>) : ', a > 10) # a > 10 => 10 > 10 

print('Greater than or Equal (>=) : ', a >= 10) # a >= 10 => 10 >= 10 

print('Greater than or Equal (>=) : ', a >= b) # a >= b => 10 >= 5 

print('Less than (<) : ', a < 10) # a < 10 => 10 < 10 

print('Less than or Equal (<=) : ', a <= 10) # a <= 10 => 10 <= 10 

print('Less than or Equal (<=) : ', a <= b) # a <= b => 10 <= 5 
```

# 4. Logical Operators

```python
userName = input('Enter your User Name : ')

password = input('Enter your Password : ')

print(userName == 'admin')

print(password == 'admin@123')

print('--------------------------------------')

# Logical And Operators
'''
Any One Condition have False. Then Everything is False
'''

print((userName == 'admin') and (password == 'admin@123'))

print('--------------------------------------')

# Logical Or Operators
'''
Any One Condition have True. Then Everything is True
'''

print((userName == 'admin') or (password == 'admin@123'))

print('--------------------------------------')

# Logical Not Operators
'''
one or more Condition have True. Then Return False

one or more Condition have False. Then Return True
'''

print(not(((userName == 'admin') and (password == 'admin@123'))))
```

# 5. Membership Operators

```python
details = [101, 'Dhara', 27, 'Female']

print(102 in details)

print(101 in details)

print('-------------------------------------------')

print(102 not in details)

print(101 not in details)
```

# 6. Identity Operators

```python
a = 10

b = 10

print(id(a))

print(id(b))

print(a is b)

print('-----------------------------------------')

print(a is not b)
```

# 7. Bitwise Operators

```python
'''

a = 10 = 1010

b = 2  = 0010 # 1 = True, 0 = False, Any one have False then Everything is False
-------------
         0010
            
'''

a = 10

b = 2

print('Bitwise and Operator(&) : ', a & b)
```

```python
'''

a = 10 = 1010

b = 2  = 0010 # 1 = True, 0 = False, Any one have True then Everything is True
-------------
         1010
            
'''

a = 10

b = 2

print('Bitwise or Operator(|) : ', a | b)
```

```python
'''

a = 10 = 1010

b = 2  = 0010 # 1 = True, 0 = False, Both are True, False. Then Return False, Any one have True. Then Averything is true
-------------
         1000
            
'''
 
a = 10

b = 2

print('Bitwise XOR Operator(^) : ', a ^ b)
```

```python
'''
a = 10
~a  = -(a + 1)
    = -(10 + 1)
    = -11
'''

a = 10

print('Bitwise Not Operators (~) : ', ~a)
```

```python
# Left Shift Operators

'''
a << 3 = 10 * (2 * 2 * 2)
       = 10 * (8)
       = 80
'''

a = 10

print('Left Shift Operators (<<) : ', a << 2)

print('Left Shift Operators (<<) : ', a << 3)
```

```python
# Right Shift Operators

'''
a >> 3 = 10 / (2 * 2 * 2)
       = 10 / (8)
       = 80
'''

a = 10

print('Right Shift Operators (>>) : ', a >> 2)

print('Right Shift Operators (>>) : ', a >> 3)
```
