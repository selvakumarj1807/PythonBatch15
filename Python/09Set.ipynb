# Set
-----

1. Tuple is a collection of Data Type.
2. declare - { }
3. Set are used to store many values & Different Data Type Values in a single Variable. Based On Non Index.
4. Set are Mutable(Changable)
5. Set are Used for Remove Douplicate Values

```python
userData = {101, 'Ram', 'male', 101, 'Ram'}

print(userData, type(userData))
```

# Add Values

```python
userData = {101, 'Ram', 'male', 101, 'Ram'}

print(userData)

print('--------------------------------------------------')

userData.add('TN')

print(userData)
```

# Update

```python
set01 = {1, 2, 3, 4, 1, 2}

set02 = {1, 2, 5, 6, 7, 5, 6}

print('--------------------------------')

set01.update(set02)

print(set01)
```

# Rmove values using Remove Method

```python
userData = {101, 'Ram', 'male', 101, 'Ram'}

print(userData)

print('--------------------------------------------------')

userData.remove(101)

print(userData)
```

# Remove values using Discard Method

```python
userData = {101, 'Ram', 'male', 101, 'Ram'}

print(userData)

print('--------------------------------------------------')

userData.discard(101)

print(userData)
```

# Difference between Remove & Discard

```python
# Using Remove

userData = {101, 'Ram', 'male', 101, 'Ram'}

print(userData)

print('--------------------------------------------------')

userData.remove(101)

print(userData)

print('--------------------------------------------------')

userData.remove('Ram kumar')

print(userData)
```

```python
# Using Discard

userData = {101, 'Ram', 'male', 101, 'Ram'}

print(userData)

print('--------------------------------------------------')

userData.discard(101)

print(userData)

print('--------------------------------------------------')

userData.discard('Ram kumar')

print(userData)
```

# Union

```python
set01 = {1, 2, 3, 4, 1, 2}

set02 = {1, 2, 5, 6, 7, 5, 6}

print('--------------------------------')

r1 = set01.union(set02)

print(r1)
```

# Intersection

```python
set01 = {1, 2, 3, 4, 1, 2}

set02 = {1, 2, 5, 6, 7, 5, 6}

print('--------------------------------')

r1 = set01.intersection(set02)

print(r1)
```

# Symmetric Difference

```python
set01 = {1, 2, 3, 4, 1, 2}

set02 = {1, 2, 5, 6, 7, 5, 6}

print('--------------------------------')

r1 = set01.symmetric_difference(set02)

print(r1)
```
