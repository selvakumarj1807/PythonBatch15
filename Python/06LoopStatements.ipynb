# While Loop

```python
i = 0 # Value Initial

n = 10

while i <= n: # 0 <= 10 # Value Condition
    print(i)

    i += 1 # Value Increment
    
```

# To Print the Even Numbers

```python
i = 0 

n = 10

while i <= n: # 4 <= 10
    print(i)

    i += 2 
    
```

# To Print the Odd Number

```python
i = 1 

n = 10

while i <= n: # 4 <= 10
    print(i)

    i += 2 
    
```

# Jumping Statements

# Break

```python
i = 0 

n = 10

while i <= n: 
    print(i)

    if i == 5:
        print('Before Break Statement')
        break
        print('After Break Statement')

    i += 1 
    
```

# Continue

```python
i = 0 

n = 10

while i <= n: # 7 <= 10
    print(i)

    if i == 5:
        i += 2
        print('Before Continue Statement')
        continue
        print('After Continue Statement')

    i += 1 
    
```

# Pass

```python
i = 0 

n = 10

while i <= n: # 8 <= 10
    print(i)

    if i == 5:
        i += 2
        print('Before Pass Statement')
        pass
        print('After Pass Statement')

    i += 1 
    
```

# While Loop

```python
i = 0 # Value Initial

n = 10

while i <= n: # Value Conditional # 11 <= 10
    print(i)

    i = i + 1 # Value Increment
```

# For Loop

1. Default Value Initial is 0
2. Default Value Increment is 1

```python
for i in range(10+1):
    print(i)
```

```python
for i in range(1, 10):
    print(i)
```

# To Print Odd Number

```python
# (Value Initial, Value Condition, Value Increment)

for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(0, 10, 2):
    print(i)
```
