# Concatination

```python
# Variable

a = 10

b = 5

c = a + b

print('The Sum Of Two Number is : ', a)

print(a, ' + ', b, ' = ', c) # 10 + 5 = 15

print(f"{a} + {b} = {c}")
```

```python
name = 'Dharani'

age = 27

gender = 'Female'

print(name, gender)

print(name + ' ' + gender)

print(name , age)
```

# Type Casting

```python
v1 = 27

v2 = 25.55

print(v1, type(v1))

print(v2, type(v2))

print('-----------------------------------')

v1 = float(v1)

v2 = int(v2)

print(v1, type(v1))

print(v2, type(v2))
```

# The area of a circle is calculated using the formula  or  A = pi * r * r

```python
pi = 3.14

r = int(input('Enter your Radius Amount : '))

A = pi * r * r

print('The area of a circle for radius of', r, 'is', A )
```
