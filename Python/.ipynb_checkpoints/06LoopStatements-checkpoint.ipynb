# While Loop

```python
i = 0 # Value Initial

n = 10

while i <= n: # 0 <= 10 # Value Condition
    print(i)

    i += 1 # Value Increment
    
```

# To Print the Even Numbers

```python
i = 0 

n = 10

while i <= n: # 4 <= 10
    print(i)

    i += 2 
    
```

# To Print the Odd Number

```python
i = 1 

n = 10

while i <= n: # 4 <= 10
    print(i)

    i += 2 
    
```

# Jumping Statements

# Break

```python
i = 0 

n = 10

while i <= n: 
    print(i)

    if i == 5:
        print('Before Break Statement')
        break
        print('After Break Statement')

    i += 1 
    
```

# Continue

```python
i = 0 

n = 10

while i <= n: # 7 <= 10
    print(i)

    if i == 5:
        i += 2
        print('Before Continue Statement')
        continue
        print('After Continue Statement')

    i += 1 
    
```

# Pass

```python
i = 0 

n = 10

while i <= n: # 8 <= 10
    print(i)

    if i == 5:
        i += 2
        print('Before Pass Statement')
        pass
        print('After Pass Statement')

    i += 1 
    
```
