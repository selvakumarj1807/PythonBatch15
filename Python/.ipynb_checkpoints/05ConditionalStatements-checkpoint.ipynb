# IF

```python
age = int(input('Enter your Age Values : '))

print(age >= 18)

print('-----------------------------------')

if age >= 18:
    print(age, 'is above 18. So you are eligible for vote!..')
```

# If & Else

```python
age = int(input('Enter your Age Values : '))

print(age >= 18)

print('-----------------------------------')

if age >= 18:
    print(age, 'is above 18. So you are eligible for vote!..')

else:
    print(age, 'is below 18. So you are not eligible for vote!..')
```

# Conditional Statement using Logical Operators

```python
username = input('Enter your UserName : ')

password = input('Enter your Password : ')

print(username == 'admin')

print(password == 'admin@123')

print('--------------------------------------')

if ((username == 'admin') and (password == 'admin@123')):
    print('Welcome Admin!...')

else:
    print('Invalid Credentials!...')
```

# If & Elif & Else

```python
username = input('Enter your UserName : ')

password = input('Enter your Password : ')

print('------------------------------------------------------')

if ((username == 'admin') and (password == 'admin@123')):
    print('Welcome Admin!...')

elif ((username == 'user01') and (password == 'user01@123')):
    print('Welcome User01!...')

elif ((username == 'user02') and (password == 'user02@123')):
    print('Welcome User02!...')

else:
    print('Invalid Credentials!...')
```

# Nested If

```python
m1 = int(input('Enter Mark 01: '))

m2 = int(input('Enter Mark 02: '))

print('--------------------------------------------')

if m1 >= 35 and m2 >= 35:
    total = m1 + m2
    average = int(total / 2)

    print('Total : ', total)

    print('Average : ', average)

    if average >= 90 and average <= 100:
        print(average, '- A Grade')

    elif average >= 80 and average <= 90:
        print(average, '- B Grade')
        
    elif average >= 70 and average <= 80:
        print(average, '- C Grade')

    else:
        print(average, '- No Grade')
    

else:
    print('You are Fail')

    if m1 <= 35:
        print(m1, 'is Below 35. So you are Fail')
        
    if m2 <= 35:
        print(m2, 'is Below 35. So you are Fail')
```
