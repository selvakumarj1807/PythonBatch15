# List
------

1. List is a collection of Data Type.
2. declare - [ ]
3. List are used to store many values & Different Data Type Values in a single Variable. Based On Index. Index start position is 0
4. List are Mutable(Changable)

```python
data = [101, 'Sam', 27, 95.5, True]

print(data, type(data))
```

# Select List Values

```python
print(data[1])

# variableName[startPosition : endPosition + 1]

print(data[1:4])
```

```python
print(data[-4])

# variableName[startPosition : endPosition + 1]

print(data[-4:-1])
```

```python
print(data[:3])

print(data[2:])
```

# Add a Single Value Using Append

```python
data = [101, 'Sam', 27, 95.5, True]

print(data)

print('-------------------------------------------------')

data.append('Male')

print(data)
```

# Add a Multiple Value Using Extend

```python
data01 = [101, 'Sam', 27, 95.5, True]

data02 = ['male', 'TN', 'CHE']

print('-------------------------------------------------')

data01.extend(data02)

print(data01)
```

# Update Value

```python
data = [101, 'Sam', 27, 95.5, True]

print(data)

print('-------------------------------------------------')

print(data[1])

data[1] = 'Ram'

print(data)
```

# Insert Values

```python
data = [101, 'Sam', 27, 95.5, True]

print(data)

print('-------------------------------------------------')

data.insert(3, 'Male')

print(data)
```

# Delete Values using Remove Method

```python
data = [101, 'Sam', 27, 95.5, True]

print(data)

print('-------------------------------------------------')

data.remove('Sam')

print(data)
```

# Delete Values using Pop Method. Based on Index Position

```python
data = [101, 'Sam', 27, 95.5, True]

print(data)

print('-------------------------------------------------')

data.pop(1)

print(data)
```

# Sorting (Assending or Descending)

```python
ageValues = [20, 18, 25, 22, 21, 30, 29, 17, 33, 22]

print(ageValues)

print('-----------------------------------------------')

ageValues.sort()

print(ageValues)

print('-----------------------------------------------')

ageValues.sort(reverse=True)

print(ageValues)
```

```python
data = [101, 'Sam', 27, 95.5, True]

print(data)

print('-------------------------------------------------')

print('Sam' in data)
```

```python
indx = 1

for i in data:
    print(indx, ' : ', i)

    indx += 1
```
